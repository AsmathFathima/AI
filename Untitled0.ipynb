```python
import numpy as np
x=np.random.uniform(-10,10)
eeta=0.005
epochs=200
iteration=0
while iteration<epochs:
  slope=4*(x**3)+6*x
  step=(-eeta)*slope
  x=x+step
  iteration=iteration+1
  x=round(x)
  obj=x**4+3*(x**2)+10
  print('x=',x)
  print('global minimum point=',obj)
```
